#### "Mini Project: 퍼셉트론으로 분류" 슬라이드의 문제를 풀어서 제출하세요.
#### Weights(가중치)가 정의 되어야 하며, 해당 데이터들이 퍼셉트론에 의해 분류가 됨을 보여줘야 합니다. 


````python
import numpy as np

epsilon = 0.0000001		# 부동소수점 오차 방지

def step_func(t):		# 퍼셉트론의 활성화 함수
    if t > epsilon: return 1
    else: return 0

def perceptron_fit(X, Y, epochs=15):	# 퍼셉트론 학습 알고리즘 구현
    global W
    eta = 0.1				# 학습률
    for t in range(epochs):
        print("epoch=", t, "======================")
        for i in range(len(X)):
            predict = step_func(np.dot(X[i], W))
            error = Y[i] - predict		               # 오차 계산
            W += eta * error * X[i]		              # 가중치 업데이트
            print("현재 처리 입력=",X[i],"정답=",Y[i],"출력=",predict,"변경된 가중치=", W)

def perceptron_predict(X, Y):		# 예측	
    global W
    for x in X:
         print(x[0], x[1], "->", step_func(np.dot(x, W)))


# 샘플과 레이블 
X = [[160,55],[163,43],[165,48],[170,80],[175, 76],[180, 70]]
y = [0, 0, 0, 1,1,1] 

# 정규화 & bias 추가
for i in X:
  i[0] = i[0]/200
  i[1] = i[1]/100
  i.append(1)

X = np.array(X)
y = np.array(y)
W = np.zeros(len(X[0]))	# 가중치를 저장하는 넘파이 행렬

perceptron_fit(X, y)
perceptron_predict(X, y)
````
